```python
# 2. load a data set using pandas

import pandas as pd
Sales_data= {
    'Date': ['2025-10-05', '2025-10-06', '2025-10-07'],
    'Product': ["Gravel 01", "Gravel 02", "Gravel 00"],
    'Quantity': [10, 14, 12],
    'Price': [2300, 2800, 3200],
    'Totalsales': [23000, 32900, 38400]
      }
df=pd.DataFrame(Sales_data)
print(df)
```

```python
# 3. Display the first few raws of the dataset

print(df.head(1))
```

```python
# 4. Exploring the Structure of a Dataset

print(df.info())
print(df.isnull().sum())
```

```python
# 5. Cleaning the Dataset

df= df.dropna()
df= df.fillna(0)
```

```python
#  TASK TWO


# 1. Computing the Basic Stastics

# Mean for Numerical Columns

df["Totalsales"].mean()
```

```python
#  Median for Numerical Columns

df["Totalsales"].median()
```

```python
# std for Numerical Columns

df["Totalsales"].std()
```

```python
# 2. Performing Grouping

grouped = df.groupby('Product')[['Quantity','Price','Totalsales']].agg(['mean'])
print(grouped)
```

```python
# Finding from the 
# Gravel 00 generate revenue than others, because it's Totalsale mean is greater.
```

```python
# TASK THREE
# 1. line Chart

import pandas as pd
import matplotlib.pyplot as plt

plt.plot(df['Date'],df['Totalsales'], marker='o')
plt.title('Sales in Theree Days')
plt.xlabel('Date')
plt.ylabel('Totalsales')
plt.grid(True)
plt.show()
```

```python
# 2. Barchart

plt.bar(df['Product'],df['Totalsales'], color='skyblue')
plt.title('Total Sales of Gravel Product')
plt.xlabel('Product')
plt.ylabel('Totalsales')
plt.show()
```

```python
# 3. Histogram

plt.hist(df['Totalsales'], bins=5, color='orange', edgecolor='black')
plt.title('Distribution of Total Sales')
plt.xlabel('Totalsales')
plt.ylabel('frequency')
plt.show()
```

```python
4. Scatter plot

plt.scatter(df['Quantity'], df['Totalsales'], color='green')
plt.title('Quantity and Total Sales Relationship')
plt.xlabel('Quantity')
plt.ylabel('Totalsales')
plt.grid(True)
plt.show()
```
